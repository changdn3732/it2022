##  Geometric Transformations of Images


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

####    Scaling

```python
img = cv2.imread('messi.jpg')
res = cv2.resize(img,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
#OR
height, width = img.shape[:2]
res = cv2.resize(img,(2*width, 2*height), interpolation = cv2.INTER_CUBIC)

img = cv2.imwrite('messire.jpg',res)

```

####    Translation

```python
img = cv2.imread('messire.jpg',0)
rows,cols = img.shape
M = np.float32([[1,0,150],[0,1,50]])
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('img',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

####    Rotation

```python
img = cv2.imread('messire.jpg',0)
rows,cols = img.shape
M = cv2.getRotationMatrix2D((cols/2,rows/2),45,1)
dst = cv2.warpAffine(img,M,(cols,rows))
cv2.imshow('rotation25',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

####    Affine Transformation

```python
img = cv2.imread('messire.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[50,50],[200,50],[50,200]])
pts2 = np.float32([[25,75],[175,125],[100,250]])
M = cv2.getAffineTransform(pts1,pts2)
dst = cv2.warpAffine(img,M,(cols,rows))

plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

####    Perspective Transformation


```python
img = cv2.imread('paper.jpg')
rows,cols,ch = img.shape
pts1 = np.float32([[164,254],[636,313],[572,1281],[930,990]])
pts2 = np.float32([[300,300],[0,300],[300,0],[0,0]])
M = cv2.getPerspectiveTransform(pts1,pts2)
dst = cv2.warpPerspective(img,M,(300,300))
plt.subplot(121),plt.imshow(img),plt.title('Input')
plt.subplot(122),plt.imshow(dst),plt.title('Output')
plt.show()
```

##  Smoothing Images

```python
img2 = cv2.imread('lenna.png')
b, g, r = cv2.split(img2) 
img = cv2.merge([r,g,b])


kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)


plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('Averaging')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img2 = cv2.imread('opencvlogo.png')
b, g, r = cv2.split(img2) 
img = cv2.merge([r,g,b])




blur = cv2.blur(img,(5,5))


plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur),plt.title('blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img2 = cv2.imread('lenna.png')
b, g, r = cv2.split(img2) 
img = cv2.merge([r,g,b])


kernel = np.array([[1,-1,1],[-1,5,-1],[1,-1,1]],np.float32)
dst = cv2.filter2D(img,-1,kernel)


plt.subplot(121),plt.imshow(img),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst),plt.title('kennel')
plt.xticks([]), plt.yticks([])
plt.show()
```
